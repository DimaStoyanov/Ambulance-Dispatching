#  3 Госпиталя
Рассмотрим двухмерную модель с 3 госпиталями расположенных внутри окружности.   




Формулы для вероятности очереди будут такими же, как и в одномерной модели. Нужно модифицировать вычисления потока пациентов в госпитали и время их транспортировки.


### Поток пациентов

#### AAA strategy

$\lambda_i=\frac{\lambda}{3}$
#### RAA/ARA/RRR strategy

$ strategy[i]=R, strategy[j]=A $

$\begin{cases} 
\lambda_{e_i} = \lambda_i \cdot (1-p_{rej}(\lambda_i)) \\
\lambda_{e_j} = \lambda_j + \frac{\lambda_i \cdot p_{rej}(\lambda_i)}{2}
\end{cases}$

Если близжайший госпиталь не может принять пациента, он будет доставлен в следующий по близости госпиталь. В виду расположения оставшихся госпиталей, пациенты будут равномерно распределены между двумя оставшимися госпиталями.

#### ARR/RAR/RRA strategy

$ strategy[i]=R, strategy[j]=R, strategy[k]=A $

$ \lambda_m^*=\sum_{n=1}^3\lambda_{m,n} $ 

$ 
\lambda_i^{R1} = \lambda_i * p_{rej}(\lambda_i) $

$ \lambda_i^{R2} = \frac{\lambda_j^{R1}}{2} \cdot p_{rej}(\lambda_i + \frac{\lambda_j^{R1}}{2}) 
 $

$ 
  \begin{equation} 
     \begin{split}
\begin{cases}
\lambda_{i,1} = \lambda_i - \lambda_i^{R1} \\
\lambda_{i,2} = \frac{\lambda_j^{R1}}{2} - \lambda_i^{R2} 
\end{cases} 
    \end{split}
    \quad\quad\quad
    \begin{split}
\begin{cases}
\lambda_{k,1} = \lambda_k \\
\lambda_{k,2} = \frac{\lambda_i^{R1}+\lambda_j^{R1}}{2} \\
\lambda_{k,3} = \lambda_i^{R2}+\lambda_j^{R2}
\end{cases}
\end{split}
\end{equation} $

В случае двух госпиталей со стратегией R и одним со стратегией A будет 3 "потока" пациентов:
* Поток, доставленный до близжайшего госпиталя (либо госпиталь со стратгией А либо не переполненный)
* Поток, доставленный до второго по близости госпиталя (либо госпиталь со стратегией А либо не переполненный)
* Оставшиеся пациенты доставлены до госпиталя со стратегией А

#### RRR strategy

$ \lambda_i^{R1} = \lambda_i \cdot p_{rej}(\lambda_i)  $

$\lambda_i^{R2} = \frac{\lambda_j^{R1} + \lambda_k^{R1}}{2} \cdot p_{rej}(\lambda_i + \frac{\lambda_j^{R1} + \lambda_k^{R1}}{2}) $

$\lambda_i^{R3} = \frac{\lambda_j^{R2} + \lambda_k^{R2}}{2} \cdot p_{rej}(\lambda_i + \frac{\lambda_j^{R1}+\lambda_k^{R1}}{2}+\frac{\lambda_j^{R2}+\lambda_k^{R2}}{2}) $

$ \lambda_i^* = \sum_{n=1}^4 $

$ \begin{cases}
\lambda_{i,1} = \lambda_i - \lambda_i^{R1} \\
\lambda_{i,2} = \frac{\lambda_j^{R1}+\lambda_k^{R1}}{2}-\lambda_i^{R2} \\
\lambda_{i,3} = \frac{\lambda_j^{R2}+\lambda_k^{R2}}{2}-\lambda_i^{R3} \\
\lambda_{i,4} = \frac{\lambda_i^{R3}+\lambda_j^{R3}+\lambda_k^{R3}}{3}
\end{cases} $

В случае 3 госпиталей со стратегией R, будет 4 "потока" пациентов:
* Поток, доставленный до близжайшего госпиталя (если он не переполнен)
* Поток, доставленный до второго по близости госпиталя (если он не переполнен)
* Поток, доставленный до третьего по близости госпиталя (если он не переполнен)
* Оставшиеся пациенты равномерно распределены между всеми госпиталями
